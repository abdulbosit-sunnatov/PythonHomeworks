```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
def f(x):
    return x**2 - 4*x + 4

x_values = np.linspace(-10, 10, 400)
y_values = f(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label=r"$f(x) = x^2 - 4x + 4$", color="blue")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Plot of $f(x) = x^2 - 4x + 4$")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(True, linestyle="--", linewidth=0.5)

plt.show()
```

```python
#2

x_values = np.linspace(0, 2*np.pi, 400)

y_sin = np.sin(x_values)
y_cos = np.cos(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_sin, linestyle='-', color='blue', label=r'$\sin(x)$')
plt.plot(x_values, y_cos, linestyle='--', color='red', label=r'$\cos(x)$')

plt.xlabel("x-axis (radians)")
plt.ylabel("y-axis")
plt.title("Plot of $\\sin(x)$ and $\\cos(x)$")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(True, linestyle="--", linewidth=0.5)

plt.show()
```

```python
#3

x1 = np.linspace(-2, 2, 400)  
x2 = np.linspace(0, 2*np.pi, 400)  
x3 = np.linspace(0, 2, 400)  
x4 = np.linspace(0, 5, 400)  

y1 = x1**3
y2 = np.sin(x2)
y3 = np.exp(x3)
y4 = np.log(x4 + 1)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x1, y1, color='blue')
axes[0, 0].set_title(r'$f(x) = x^3$')
axes[0, 0].set_xlabel("x-axis")
axes[0, 0].set_ylabel("y-axis")
axes[0, 0].grid(True, linestyle="--", linewidth=0.5)

axes[0, 1].plot(x2, y2, color='red')
axes[0, 1].set_title(r'$f(x) = \sin(x)$')
axes[0, 1].set_xlabel("x-axis")
axes[0, 1].set_ylabel("y-axis")
axes[0, 1].grid(True, linestyle="--", linewidth=0.5)

axes[1, 0].plot(x3, y3, color='green')
axes[1, 0].set_title(r'$f(x) = e^x$')
axes[1, 0].set_xlabel("x-axis")
axes[1, 0].set_ylabel("y-axis")
axes[1, 0].grid(True, linestyle="--", linewidth=0.5)

axes[1, 1].plot(x4, y4, color='purple')
axes[1, 1].set_title(r'$f(x) = \log(x+1)$')
axes[1, 1].set_xlabel("x-axis")
axes[1, 1].set_ylabel("y-axis")
axes[1, 1].grid(True, linestyle="--", linewidth=0.5)

plt.tight_layout()

plt.show()
```

```python
np.random.seed(42)  
x_values = np.random.uniform(0, 10, 100)
y_values = np.random.uniform(0, 10, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, color='purple', marker='o', alpha=0.7, edgecolors='black')

plt.xlabel("X-axis (Random Values)")
plt.ylabel("Y-axis (Random Values)")
plt.title("Scatter Plot of 100 Random Points")
plt.grid(True, linestyle="--", linewidth=0.5)

plt.show()
```

```python
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Data (=0, =1)")
plt.grid(True, linestyle="--", linewidth=0.5)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.cos(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

fig.colorbar(surf)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot of f(x, y) = cos(x + y)')

plt.show()
```

```python
categories = np.array(['Product A', 'Product B', 'Product C', 'Product D', 'Product E'])
values = np.array([200, 150, 250, 175, 225])

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='blue', label="Sales Numbers")  # Added label
plt.title('Sales data for five different products')
plt.ylabel('Sales numbers')
plt.xlabel('Categories')
plt.legend() 
plt.show()
```

```python
time_periods = np.array(['T1', 'T2', 'T3', 'T4'])

category_A = np.random.randint(10,50,4)
category_B = np.random.randint(10,50,4)
category_C = np.random.randint(10,50,4) 

plt.bar(time_periods,category_A,color='green', label='Category A')
plt.bar(time_periods,category_B,bottom=category_A, color='blue', label='Category B')
plt.bar(time_periods,category_C,bottom=category_A+category_B,color='red', label='Category C')

plt.title('Stacked Bar Chart: Category Contributions Over Time')
plt.xlabel('Time Periods')
plt.ylabel('Values')
plt.legend()

plt.show()
```
