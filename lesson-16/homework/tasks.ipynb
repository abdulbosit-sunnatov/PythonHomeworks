```python
sample_data = {
    "FlightNumber": [101, 102, 103, 104],
    "Origin": ["New York", "Los Angeles", "Chicago", "Houston"],
    "Destination": ["London", "Tokyo", "Paris", "Berlin"],
    "Duration": [415, 720, 395, 510]
}
sample_flights_df = pd.DataFrame(sample_data)

sample_flights_df.to_parquet('data/flights.parquet', engine='pyarrow', index=False)

flights_df = pd.read_parquet('data/flights.parquet')

flights_df.info()
flights_df
```
